<h1> Merge Dataset

```python
compiled_dataset = mortgage_data.merge(consumer_price_index,how='inner',on='DATE')
compiled_dataset = compiled_dataset.merge(industrial_production,how='inner',on='DATE')
compiled_dataset = compiled_dataset.merge(consumer_goods,how='inner',on='DATE')
compiled_dataset = compiled_dataset.merge(rates_90day,how='inner',on='DATE')
compiled_dataset = compiled_dataset.merge(producer_price_index,how='inner',on='DATE')
compiled_dataset = compiled_dataset.merge(unemployment_rate,how='inner',on='DATE')
compiled_dataset['3monthForecast'] = compiled_dataset['MORTGAGE30US'].shift(-3)
compiled_dataset['6monthForecast'] = compiled_dataset['MORTGAGE30US'].shift(-6)
compiled_dataset['12monthForecast'] = compiled_dataset['MORTGAGE30US'].shift(-12)
compiled_dataset = compiled_dataset.dropna()
compiled_dataset
```

```python
compiled_dataset = compiled_dataset.merge(money_supply,how='left',on='DATE')
compiled_dataset
```

```python
compiled_dataset.info()
```

<h1> Complete Case Analysis (CCA)

```python
cca_set = compiled_dataset.dropna()
cca_set.info()
```

```python
#Remove date and response columns
X = cca_set.drop(['DATE','MORTGAGE30US','3monthForecast', '6monthForecast', '12monthForecast'], axis = 1)
y = cca_set['3monthForecast']

#Create training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2023)

#Train forecast model
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)

#Test 3 month forecast
y_predict = regressor.predict(X_test)

#Evaluate Model
slope, intercept = np.polyfit(y_test, y_predict, 1)
plt.scatter(y_test, y_predict, color = 'gray')
plt.plot(y_test, slope * y_test + intercept, color='red')
plt.ylabel('Model Predictions')
plt.xlabel('True Values')
plt.title('CCA Analysis [80% Training Model ~400 training points]')

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)), '.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
print('RSME = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('R2 = ', r2)
```

<h1> Impute Missing Data

```python
#Remove date and response columns
cleaned_data = compiled_dataset.drop(['DATE','MORTGAGE30US','3monthForecast', '6monthForecast', '12monthForecast'], axis = 1)
column_names = cleaned_data.columns.tolist()

#Impute using linear regression
lr = LinearRegression()
imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=10, verbose=2, imputation_order='roman',random_state=0)
cleaned_data = imp.fit_transform(cleaned_data)
```

```python
cleaned_data = pd.DataFrame(cleaned_data, columns=column_names)
cleaned_data.head(20)
```

<h4> Linear Imputation produces negative values for money supply. Negative values for U.S Money Supply is not a realistic value for predicting mortgage rates. Alternative imputation methods must be considered.

```python
#Remove date and response columns
cleaned_data = compiled_dataset.drop(['DATE','MORTGAGE30US','3monthForecast', '6monthForecast', '12monthForecast'], axis = 1)

#Impute using KNN
knn_imputer = KNNImputer(n_neighbors=10)
cleaned_data = knn_imputer.fit_transform(cleaned_data)

cleaned_data = pd.DataFrame(cleaned_data, columns=column_names)
cleaned_data
```

<h1> Analyze Imputed Data

```python
X = cleaned_data
y = compiled_dataset['3monthForecast']

#Create training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2023)

#Train forecast model
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)

#Test 3 month forecast
y_predict = regressor.predict(X_test)

#Evaluate Model
slope, intercept = np.polyfit(y_test, y_predict, 1)
plt.scatter(y_test, y_predict, color = 'gray')
plt.plot(y_test, slope * y_test + intercept, color='red')
plt.ylabel('Model Predictions')
plt.xlabel('True Values')
plt.title('Data Imputation Analysis [80% Training Model ~500 training points]')

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)), '.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
print('RSME = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('R2 = ', r2)
```
